### Exercise 1

1. Open the CSV file.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/titanic/titanic3.csv')
df
```

_____
2. Get the names of all the columns, then iterate that object and print the following row by row: "Null values in Survived: 10" where 10 is the total of null or missing values in the Survived column.

```python
column = df.columns.values

for i in column:    
    total_null = df[i].isnull().sum()
    print(f'Null values in {i}: {total_null}')
```

_____
3.Find out what the average age was and print "The average age was: mean"

```python
age_mean = round(df['age'].mean(), 2) 
print(f'The average age was: {age_mean}')
```

_____
4. Create a table with the number of survivors divided by sex.

```python
survivors = df.pivot_table(index='sex', aggfunc={'survived': sum})
survivors
```

```python
df_2.eval('`Wind+Solar`= Wind + Solar', inplace=True) 
df_2
```

```python
df_2.drop('Wind+Solar', axis=1, inplace=True) 
df_2
```

```python
df_2.rename(columns={'BACKTICK_QUOTED_STRING_Wind_PLUS_Solar':'Wind+Solar'}, inplace=True)
df_2
```

4. Create a pivot table that displays Wind + Solar consumption according to date, it must be vertical.

```python
# check if there are null values
df_2.Consumption.isnull().sum()
```

```python
table = df_2.pivot_table('Consumption', index='Date')
table
```
